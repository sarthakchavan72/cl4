```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Load the iris dataset
iris = load_iris()
X = iris.data  # Features

# Initialize KMeans with the number of clusters and n_init set explicitly to avoid warning
kmeans = KMeans(n_clusters=3, n_init='auto', random_state=42)

# Fit KMeans to the data
kmeans.fit(X)

# Get cluster centroids and labels
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

# Visualizing the clusters (using first two features for 2D plot)
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', label='Data points')
plt.scatter(centroids[:, 0], centroids[:, 1], marker='X', s=200, c='red', label='Centroids')
plt.title('K-Means Clustering on Iris Dataset')
plt.xlabel('Feature 1: Sepal length (cm)')
plt.ylabel('Feature 2: Sepal width (cm)')
plt.legend()
plt.grid(True)
plt.show()
```
